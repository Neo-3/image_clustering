```python
from keras.preprocessing import image
from keras.applications.vgg16 import VGG16
from keras.applications.vgg16 import preprocess_input
import numpy as np
from sklearn.cluster import KMeans
import os, shutil, glob, os.path
from PIL import Image as pil_image
import matplotlib.pyplot as plt
```

```python
# load VGG16 model

image.LOAD_TRUNCATED_IMAGES = True 
model = VGG16(weights='imagenet', include_top=False)
```

```python
# set dataset folder and target folder
imgdir = './downloads'
# get images
files = glob.glob(os.path.join(imgdir, '*.jpg'))
files.sort()
```

```python
# Using Keras to get data from image

features = []

for i, imagepath in enumerate(files):
    print("Status:", i+1, "/", len(files), end="\r")
    
    img = image.load_img(imagepath, target_size=(256, 256))
    img_data = image.img_to_array(img)
    img_data = np.expand_dims(img_data, axis=0)
    img_data = preprocess_input(img_data)
    
    feat = np.array(model.predict(img_data))
    features.append(feat.flatten())
```

```python
df=np.array(features)
```

# Kmeans

```python
number_clusters = 8

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=number_clusters)
kmeans.fit(df)
y_kmeans = kmeans.predict(df)
```

```python
cluster_centers = kmeans.cluster_centers_
y_kmeans = kmeans.predict(df)
```

```python
plt.figure(figsize=(10, 7))

plt.scatter(df[:, 0], df[:, 1], c=y_kmeans, s=750, cmap='bone',alpha=0.6,edgecolors='k', marker='.')


cluster_centers = kmeans.cluster_centers_
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='darkgoldenrod',edgecolors='darkgoldenrod', s=150, alpha=0.8)

plt.show()
```

# Meanshift 

```python
from sklearn.cluster import MeanShift 

ms = MeanShift() 
ms.fit(df) 
cluster_centers = ms.cluster_centers_
y_ms = ms.predict(df)
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(df[:, 0], df[:, 1], c=y_ms, s=750, cmap='bone',alpha=0.6, marker='.')
cluster_centers = ms.cluster_centers_

plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='crimson',edgecolors='crimson', s=100, alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 7))
plt.rcParams['axes.facecolor'] = 'silver'
plt.rcParams["legend.facecolor"] = 'white'
plt.scatter(df[:, 0], df[:, 1], c=y_kmeans, s=750, cmap='gist_gray',alpha=0.7, marker='.',edgecolors='k')

centers = ms.cluster_centers_

plt.scatter(centers[:, 0], centers[:, 1], c='crimson',edgecolors='crimson', s=100, alpha=0.8,label = "meanshift")


cluster_centers = kmeans.cluster_centers_
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='y',edgecolors='y', s=100, alpha=0.8,label = "kmeans")

plt.legend()
```
